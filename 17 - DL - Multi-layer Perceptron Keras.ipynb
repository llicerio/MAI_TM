```python
#crea tu primer MLP en keras
from keras.models import Sequential
from keras.layers import Dense
import numpy
import pandas as pd
from datetime import datetime
import tensorflow as tf
import matplotlib.pyplot as plt
#fija las semillas aleatorias para la reproducibilidad
numpy.random.seed(7)
```

```python
#carga los datos
dataset = numpy.loadtxt("pima-indians-diabetes.csv", delimiter=",")
#divido en variables de entrada y de salida
X = dataset[: , 0:8]
Y = dataset[: ,8]
```

```python
#crea el modelo
model = Sequential()
model.add(Dense (12 , input_dim=8, activation = 'relu'))
model.add(Dense (8 ,activation = 'relu'))
model.add(Dense (8 ,activation = 'relu'))
model.add(Dense (12 ,activation = 'relu'))
model.add(Dense (16 ,activation = 'relu'))
model.add(Dense (1 , activation = 'sigmoid'))
```

```python
#compila el modelo
model.compile(loss='binary_crossentropy' , optimizer='adam', metrics=['accuracy'])
```

```python
#ajusta el modelo
model.fit(X,Y, epochs = 170, batch_size=10)
```

```python
#evalua el modelo
scores = model.evaluate(X,Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
datasetpredict = numpy.loadtxt("predicciones pima-indians-diabetes alumnos.csv", delimiter=",")
resultado=model.predict(datasetpredict)
print(str(resultado))
```
