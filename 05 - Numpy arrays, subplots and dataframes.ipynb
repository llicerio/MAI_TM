```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0, 100, 0.1)
r = 0.25
N0 = 1
```

```python
N = N0 * np.exp(t*r)
```

```python
N
```

```python
len(N)
```

```python
t
```

```python
len(t)
```

```python
fig, ax = plt.subplots()
ax.plot(t, N, linewidth=3)
#Establecer el título del gráfico y los ejes de la etiqueta.
ax.set_title("Población (N) através del tiempo (t)", fontsize=14)
ax.set_xlabel("Tiempo (t)", fontsize=14)
ax.set_ylabel("Población (N)", fontsize=14)
#Establecer el tamaño de las etiquetas de las marcas.
ax.tick_params(axis='both', labelsize=14)
plt.show()
```

```python
np.savetxt("valores_N.csv", N)
np.savetxt("valores_t.csv", t)
```
