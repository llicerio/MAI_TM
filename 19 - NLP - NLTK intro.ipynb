```python
import nltk
nltk.download('movie_reviews')
```

```python
from nltk.corpus import movie_reviews
```

```python
movie_reviews.categories()
```

```python
cat_neg_char = movie_reviews.raw(categories='neg')
```

```python
arreglo_char_neg=[]
```

```python
dist_frec_neg_char = nltk.FreqDist([w.lower() for w in cat_neg_char])
for m in dist_frec_neg_char:
    for i in range(dist_frec_neg_char[m]):
        arreglo_char_neg.append(m)
    print (m + ' : ' , dist_frec_neg_char[m])
```

```python
dist_frec_neg_char.most_common(20)
```

```python
import matplotlib.pyplot as plt
```

```python
from collections import Counter
import numpy as np

#counts_arreglo_char_neg = Counter(arreglo_char_neg)

counts_arreglo_char_neg = dict(Counter(arreglo_char_neg).most_common(30))

labels, values = zip(*counts_arreglo_char_neg.items())


indSort = np.argsort(values)[::-1]

labels = np.array(labels)[indSort]
values = np.array(values)[indSort]

indexes = np.arange(len(labels))

bar_width = 0.35

plt.bar(indexes, values)

plt.xticks(indexes + bar_width, labels)
plt.show()

```

```python
dist_frec_neg_char.plot(20,cumulative='True')
```

```python
import unidecode
import re

texto_prueba='café is né orčpžsíáýd'
texto_prueba_2 =  unidecode.unidecode(texto_prueba)
print (texto_prueba_2)

tp3='ella cumple años el 13 de enero, así que, ¡hoy es su cumpleaños! ¿cuantos años cumple?'
tp3_new= unidecode.unidecode(tp3)
tp3_new2 = re.sub(r'[^a-zA-Z]', ' ', tp3_new)
print(tp3_new2)

tp3_new3 =re.sub(r'^\s*|\s\s*', ' ', tp3_new2).strip()
print (tp3_new3)
```

```python
cat_neg_words = movie_reviews.words(categories='neg')
```

```python
filter_spaces=['',' ']
cat_neg_words_clean=[]
for i in cat_neg_words:
    step1=unidecode.unidecode(i)
    step2=re.sub(r'[^a-zA-Z]', '', step1)
    step3=re.sub(r'^\s*|\s\s*', '', step2).strip()
    if (step3 not in filter_spaces):        
        cat_neg_words_clean.append(step3)   
    
```

```python
dist_frec_neg_words = nltk.FreqDist([w.lower() for w in cat_neg_words_clean])
```

```python
print(dist_frec_neg_words)
```

```python
for m in dist_frec_neg_words:
    print (m + ' : ' , dist_frec_neg_words[m])
```

```python
dist_frec_neg_words.most_common(20)
```

```python
dist_frec_neg_words.plot(20,cumulative='True')
```

```python
cat_pos_words = movie_reviews.words(categories='pos')
```

```python
filter_spaces=['',' ']
cat_pos_words_clean=[]
for i in cat_pos_words:
    step1=unidecode.unidecode(i)
    step2=re.sub(r'[^a-zA-Z]', '', step1)
    step3=re.sub(r'^\s*|\s\s*', '', step2).strip()
    if (step3 not in filter_spaces):        
        cat_pos_words_clean.append(step3)   
```

```python
dist_frec_pos_words = nltk.FreqDist([w.lower() for w in cat_pos_words_clean])
```
