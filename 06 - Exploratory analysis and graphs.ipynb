```python
#Busqueda de valores nulos
print(df_diabetes.isnull().sum())
```

```python
#Busqueda de valores fuera de rango

print(df_diabetes.where(df_diabetes["Embarazos"] < 0).value_counts().sum())
print(df_diabetes.where(df_diabetes["Glucosa"] <= 0).value_counts().sum())
print(df_diabetes.where(df_diabetes["PresionArterial"] <= 0).value_counts().sum())
print(df_diabetes.where(df_diabetes["EspesorPiel"] <=0).value_counts().sum())
print(df_diabetes.where(df_diabetes["Insulina"] <= 0).value_counts().sum())
print(df_diabetes.where(df_diabetes["IndiceMasaCorporal"] <= 0).value_counts().sum())
print(df_diabetes.where(df_diabetes["GenealogiaDiabetes"] <= 0).value_counts().sum())
print(df_diabetes.where(df_diabetes["Edad"] <= 0).value_counts().sum())

```

```python
#Remplazar valores fuera de rango con NaN en el dataframe

columnasfuerarango = ["Glucosa", "PresionArterial", "EspesorPiel", "Insulina", "IndiceMasaCorporal"]
for columna in columnasfuerarango:
    df_diabetes[columna].replace(0,np.NaN,inplace=True)

```

```python
#Busqueda de valores nulos
df_diabetes.isnull().sum()
```

```python
#Remplazar valores nulos con la mediana de cada grupo de la variable Salida

print(df_diabetes.groupby(["Salida"])["Glucosa"].median())
df_diabetes["Glucosa"].fillna(df_diabetes.groupby(["Salida"])["Glucosa"].transform("median"),inplace=True)

print(df_diabetes.groupby(["Salida"])["PresionArterial"].median())
df_diabetes["PresionArterial"].fillna(df_diabetes.groupby(["Salida"])["PresionArterial"].transform("median"),inplace=True)

print(df_diabetes.groupby(["Salida"])["EspesorPiel"].median())
df_diabetes["EspesorPiel"].fillna(df_diabetes.groupby(["Salida"])["EspesorPiel"].transform("median"),inplace=True)

print(df_diabetes.groupby(["Salida"])["Insulina"].median())
df_diabetes["Insulina"].fillna(df_diabetes.groupby(["Salida"])["Insulina"].transform("median"),inplace=True)

print(df_diabetes.groupby(["Salida"])["IndiceMasaCorporal"].median())
df_diabetes["IndiceMasaCorporal"].fillna(df_diabetes.groupby(["Salida"])["IndiceMasaCorporal"].transform("median"),inplace=True)


```

```python
#Busqueda de valores nulos
df_diabetes.isnull().sum()
```
