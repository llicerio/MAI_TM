```python
OBSERVACIONES sobre resizing, tanto horizontal como vertical: Al aplicar los métodos de interpolación INTER_CUBIC e INTER_LINEAR,
    se observa que en la interpolación cúbica las images cambia más el color de los pixeles, con respecto a la imagen original. Con la
    interpolación lineal, es casi imperceptible el cambio. La diferencia se debe a que la interpolación cúbica hace el promedio de 
    16 pixeles para crear un nuevo pixel interpolado, mientras que la interpolación lineal solo usa 4 pixeles.
    
```

Ejecuta un desplazamiento de 50 pixeles sobre el eje horizontal de las imágenes originales.

```python
alto1, ancho1 = img01_girl.shape[:2]
tx1, ty1= 50, 0
matriz_tras1= np.array([[1,0,tx1], [0,1,ty1]], dtype=np.float32)
img01_tras=cv2.warpAffine(src= img01_girl , M = matriz_tras1, dsize =(ancho1, alto1))
plot.imshow(img01_tras)
```

```python
alto2, ancho2 = img02_snowman.shape[:2]
tx2, ty2= 50, 0
matriz_tras2= np.array([[1,0,tx2], [0,1,ty2]], dtype=np.float32)
img02_tras=cv2.warpAffine(src= img02_snowman , M = matriz_tras2, dsize =(ancho2, alto2))
plot.imshow(img02_tras)
```

Realiza una rotación de 45 grados de las imágenes originales

```python
centro1= (img01_girl.shape[1]//2,img01_girl.shape[0]//2)
angulorotacion1=cv2.getRotationMatrix2D(centro1, -45, 1)
img01_girl_rot = cv2.warpAffine(img01_girl,angulorotacion1,(img01_girl.shape[1], img01_girl.shape[0]))
plot.imshow (img01_girl_rot)
```

```python
centro2= (img02_snowman.shape[1]//2,img02_snowman.shape[0]//2)
angulorotacion2=cv2.getRotationMatrix2D(centro2, -45, 1)
img02_snowman_rot = cv2.warpAffine(img02_snowman,angulorotacion2,(img02_snowman.shape[1], img02_snowman.shape[0]))
plot.imshow (img02_snowman_rot)
```
